```python
import pandas as pd
url = '/content/Housing.csv'
data = pd.read_csv(url)
data.head()
```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot=True)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['area']),y=data['price'])
print('He so intercept la:',model.intercept_)
print('He so coeffient la: ',model.coef_)

a = model.score(X=pd.DataFrame(data['area']),y=data['price'])
print("Accuracy is",round(a*100,3),"%")

y_pred = model.predict(X=pd.DataFrame(data['area']))

data.plot(kind='scatter',x='area',y='price',figsize=(9,9),color='green', xlim=(1000,17000))
plt.plot(data['area'],y_pred,color='red')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
X=pd.DataFrame([data['area'],data['area']**2]).T
model.fit(X,y=data['price'])
print('He so intercept la:',model.intercept_)
print('He so coeffient la: ',model.coef_)

a = model.score(X,y=data['price'])
print("Accuracy is",round(a*100,3),"%")


import numpy as np
range = np.arange(1000,17000,100)
X_range = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X=X_range)
data.plot(kind ='scatter',x='area',y='price',figsize = (9,9),color='green',xlim=(1000,17000))
plt.plot(range,y_pred,color='red')
```
