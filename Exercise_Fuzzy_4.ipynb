```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl 
import random

BMI = ctrl.Antecedent(np.arange(10,35.1,0.1),'BMI')
AcLevel = ctrl.Antecedent(np.arange(1,2.5,0.01),'AcLevel')
Age = ctrl.Antecedent(np.arange(14,65,1),'Age')
Calo = ctrl.Consequent(np.arange(850,3001,1),'Calo')

Age['young'] = fuzz.trimf(Age.universe,[14,14,30])
Age['adult'] = fuzz.trimf(Age.universe,[18,30,64])
Age.view()

BMI['skinny'] = fuzz.trimf(BMI.universe,[10,10,17])
BMI['thin'] = fuzz.trimf(BMI.universe,[17,18,18.5])
BMI['normal'] = fuzz.trimf(BMI.universe,[18.5,21,25])
BMI['overweight'] = fuzz.trimf(BMI.universe,[25,26,27])
BMI['obesity'] = fuzz.trimf(BMI.universe,[27,35,35])
BMI.view()

AcLevel['verylight'] = fuzz.trimf(AcLevel.universe,[1,1.3,1.65])
AcLevel['light'] = fuzz.trimf(AcLevel.universe,[1.3,1.65,1.76])
AcLevel['moderate'] = fuzz.trimf(AcLevel.universe,[1.65,1.76,2.1])
AcLevel['heavy'] = fuzz.trimf(AcLevel.universe,[1.76,2.1,2.1])
AcLevel.view()

Calo['Low'] = fuzz.trimf(Calo.universe,[850,850,1700])
Calo['Normal'] = fuzz.trimf(Calo.universe,[1000,1700,2200])
Calo['High'] = fuzz.trimf(Calo.universe,[1700,3000,3000])

rule1 = ctrl.Rule(Age['young']& BMI['skinny']&AcLevel['verylight'],Calo['Low'])
rule2 = ctrl.Rule(Age['young']& BMI['thin']&AcLevel['verylight'],Calo['Low'])
rule3 = ctrl.Rule(Age['young']& BMI['normal']&AcLevel['verylight'],Calo['Low'])
rule4 = ctrl.Rule(Age['young']& BMI['overweight']&AcLevel['verylight'],Calo['Low'])
rule5 = ctrl.Rule(Age['young']& BMI['obesity']&AcLevel['verylight'],Calo['Low'])

rule6 = ctrl.Rule(Age['young']& BMI['skinny']&AcLevel['light'],Calo['Normal'])
rule7 = ctrl.Rule(Age['young']& BMI['thin']&AcLevel['light'],Calo['Normal'])
rule8 = ctrl.Rule(Age['young']& BMI['normal']&AcLevel['light'],Calo['Low'])
rule9 = ctrl.Rule(Age['young']& BMI['overweight']&AcLevel['light'],Calo['Low'])
rule10 = ctrl.Rule(Age['young']& BMI['obesity']&AcLevel['light'],Calo['Low'])

rule11 = ctrl.Rule(Age['young']& BMI['skinny']&AcLevel['moderate'],Calo['High'])
rule12 = ctrl.Rule(Age['young']& BMI['thin']&AcLevel['moderate'],Calo['High'])
rule13 = ctrl.Rule(Age['young']& BMI['normal']&AcLevel['moderate'],Calo['Normal'])
rule14 = ctrl.Rule(Age['young']& BMI['overweight']&AcLevel['moderate'],Calo['Normal'])
rule15 = ctrl.Rule(Age['young']& BMI['obesity']&AcLevel['moderate'],Calo['Normal'])

rule16 = ctrl.Rule(Age['young']& BMI['skinny']&AcLevel['heavy'],Calo['High'])
rule17 = ctrl.Rule(Age['young']& BMI['thin']&AcLevel['heavy'],Calo['High'])
rule18 = ctrl.Rule(Age['young']& BMI['normal']&AcLevel['heavy'],Calo['High'])
rule19 = ctrl.Rule(Age['young']& BMI['overweight']&AcLevel['heavy'],Calo['Normal'])
rule20 = ctrl.Rule(Age['young']& BMI['obesity']&AcLevel['heavy'],Calo['Normal'])

rule21 = ctrl.Rule(Age['adult']& BMI['skinny']&AcLevel['verylight'],Calo['Low'])
rule22 = ctrl.Rule(Age['adult']& BMI['thin']&AcLevel['verylight'],Calo['Low'])
rule23 = ctrl.Rule(Age['adult']& BMI['normal']&AcLevel['verylight'],Calo['Low'])
rule24 = ctrl.Rule(Age['adult']& BMI['overweight']&AcLevel['verylight'],Calo['Low'])
rule25 = ctrl.Rule(Age['adult']& BMI['obesity']&AcLevel['verylight'],Calo['Low'])

rule26 = ctrl.Rule(Age['adult']& BMI['skinny']&AcLevel['light'],Calo['Normal'])
rule27 = ctrl.Rule(Age['adult']& BMI['thin']&AcLevel['light'],Calo['Normal'])
rule28 = ctrl.Rule(Age['adult']& BMI['normal']&AcLevel['light'],Calo['Low'])
rule29 = ctrl.Rule(Age['adult']& BMI['overweight']&AcLevel['light'],Calo['Low'])
rule30 = ctrl.Rule(Age['adult']& BMI['obesity']&AcLevel['light'],Calo['Low'])

rule31 = ctrl.Rule(Age['adult']& BMI['skinny']&AcLevel['moderate'],Calo['High'])
rule32 = ctrl.Rule(Age['adult']& BMI['thin']&AcLevel['moderate'],Calo['High'])
rule33 = ctrl.Rule(Age['adult']& BMI['normal']&AcLevel['moderate'],Calo['Normal'])
rule34 = ctrl.Rule(Age['adult']& BMI['overweight']&AcLevel['moderate'],Calo['Normal'])
rule35 = ctrl.Rule(Age['adult']& BMI['obesity']&AcLevel['moderate'],Calo['Normal'])

rule36 = ctrl.Rule(Age['adult']& BMI['skinny']&AcLevel['heavy'],Calo['High'])
rule37 = ctrl.Rule(Age['adult']& BMI['thin']&AcLevel['heavy'],Calo['High'])
rule38 = ctrl.Rule(Age['adult']& BMI['normal']&AcLevel['heavy'],Calo['High'])
rule39 = ctrl.Rule(Age['adult']& BMI['overweight']&AcLevel['heavy'],Calo['Normal'])
rule40 = ctrl.Rule(Age['adult']& BMI['obesity']&AcLevel['heavy'],Calo['Normal'])




recom_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12, rule13, rule14, 
                                   rule15, rule16, rule17, rule18, rule19, rule20,rule21, rule22, rule23, rule24, rule25, rule26, rule27, 
                                   rule28, rule29, rule30, rule31, rule32, rule33, rule34, 
                                   rule35, rule36, rule37, rule38, rule39, rule40])
recom = ctrl.ControlSystemSimulation(recom_ctrl)
height = input("Height in meter: ")
weight = input("Weight in kilograms: ")
level = input("Activity level: ")
tuoi = random.randrange(14,65,1)
print('Age is:', tuoi)
recom.input['BMI'] = float(weight)/(float(height)**2)
recom.input['Age'] = int(tuoi)
recom.input['AcLevel'] = float(level)
recom.compute()
print('Need calogy is:', recom.output['Calo'])
Calo.view(sim=recom)
```
