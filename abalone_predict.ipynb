```python
import numpy as np
from sklearn import linear_model
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
url = "/content/abalone.csv"

data = pd.read_csv(url)
data["Sex"] = le.fit_transform(data['Sex'])
```

```python
data.hist()
plt.show
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.heatmap(correlation,annot=True)
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame(data['Length'])
y =   data['Diameter']
model.fit(X,y)
print('he so intercept la: ', model.intercept_)
print('he so coeffient la: ', model.coef_)            
print('do chinh xac cua mo hinh:',round(model.score(X,y)*100,7), '%') 

range = np.arange(0,1, 0.0001)
X_range = pd.DataFrame(range)
y_pred = model.predict(X= X_range)
data.plot(kind = 'scatter', x= 'Length', y ='Diameter', figsize= (9,9), color = "black", xlim = (0,1))
plt.plot(range, y_pred, color = 'green') 
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Length'],data['Length']**2]).T
y =   data['Diameter']
model.fit(X,y)
print('he so intercept la: ', model.intercept_)
print('he so coeffient la: ', model.coef_)            
print('do chinh xac cua mo hinh:',round(model.score(X,y)*100,7), '%') 

range = np.arange(0,1, 0.0001)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X= X_range)
data.plot(kind = 'scatter', x= 'Length', y ='Diameter', figsize= (9,9), color = "black", xlim = (0,1))
plt.plot(range, y_pred, color = 'green') 
```
